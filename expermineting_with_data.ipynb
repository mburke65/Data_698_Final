```python
import pandas as pd
import glob

all_files = glob.glob("data_csvs/*.csv")
li = []
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    print(df.shape)
    li.append(df)
frame = pd.concat(li, axis=0, ignore_index=True)
df= frame.drop_duplicates(subset="url")
df.shape
```
